1C) TRAINING / TEST SPLIT
-

Select the first 70 rows of Class 0 and the rest 140 rows of Class 1 as the
training set and the rest of the data as the test set.

```python
class0 = data[data['class'] ==0]
class1 = data[data['class'] ==1]

class0_training = class0.iloc[:70]
class1_training = class1.iloc[:140]

class0_test = class0.iloc[70:]
class1_test = class1.iloc[140:]

training_data = pd.concat([class0_training, class1_training])
test_data = pd.concat([class0_test,class1_test])

print(training_data.info())
print(test_data.info())

```

PART 2. MODEL FIT
----
A) KNN Classification
---

```python
!pip install scikit-learn
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, f1_score, recall_score
```

```python
x_training = training_data[['pelvic_incidence', 'pelvic_tilt', 'lumbar_lordosis_angle','sacral_slope', 'pelvic_radius', 'degree_spondylolisthesis']]
y_training = training_data['class']

x_test = test_data[['pelvic_incidence', 'pelvic_tilt', 'lumbar_lordosis_angle','sacral_slope', 'pelvic_radius', 'degree_spondylolisthesis']]
y_test = test_data['class']

k_set = list(range(208,0,-3))

#Create lists to store training and test errors across range of k 
training_errors = []
test_errors = []
euclidean_k_optimal = 1
euclidean_error_optimal = 1

#Iterate through k values
for k in k_set:
    knn = KNeighborsClassifier(n_neighbors=k, metric = 'euclidean')
    knn.fit(x_training, y_training)

    #Calculate errors on training data. We use accuracy_score and not MSE since this is a classification problem. 
    ypred_training = knn.predict(x_training)
    training_errors.append(1-accuracy_score(y_training, ypred_training))

    #Calculate errors on test data. We use accuracy_score and not MSE since this is a classification problem. 
    ypred_test = knn.predict(x_test)
    test_error = 1-accuracy_score(y_test, ypred_test)
    test_errors.append(test_error)
    if test_error < euclidean_error_optimal:
        euclidean_error_optimal = test_error
        euclidean_k_optimal = k
    

#Plot training and test errors
plt.figure(figsize=(15,8))
plt.plot(k_set, training_errors,'r', label = 'Training Error')
plt.plot(k_set, test_errors,'b',label = 'Test Error')
plt.xlabel('k value')
plt.title('Plotting errors vs. k values')
plt.legend(loc = 'upper left')
plt.grid(True)
plt.show()


    
    
```

```python
#zoom in on between 1 and 10. 
plt.figure(figsize=(15,8))
plt.plot(k_set, training_errors,'r', label = 'Training Error')
plt.plot(k_set, test_errors,'b',label = 'Test Error')
plt.xlabel('k value')
plt.xlim(1,10)
plt.title('Plotting errors vs. k values')
plt.legend(loc = 'upper left')
plt.grid(True)
plt.show()
```

k* = 4 is more suitable among these values as it minimizes the test error. 
-

```python
#Fit model with k* = 4

knn_optimal = KNeighborsClassifier(n_neighbors = euclidean_k_optimal, metric = 'euclidean')
knn_optimal.fit(x_training, y_training)
ypred_test_optimal = knn_optimal.predict(x_test)
ypred_training_optimal = knn_optimal.predict(x_training)
training_error_optimal = 1 - accuracy_score(y_training, ypred_training_optimal)
test_error_optimal = 1 - accuracy_score(y_test,ypred_test_optimal)

print('Test error at k = 4: {:.2f}'.format(test_error_optimal))
print('Training error at k = 4: {:.2f}'.format(training_error_optimal))
                                


#Confusion matrix: 
cmatrix = confusion_matrix(y_test, ypred_test_optimal)
print('\nConfusion Matrix at k = 4:')
print(cmatrix)

cmatrix_flattened = cmatrix.flatten()
TN, FP, FN, TP = cmatrix_flattened

#True positive rate
TPR = TP / (TP+FN)
print('True positive rate at k =4: {:.2f}'.format(TPR))

#True negative rate
TNR = TN / (TN+FP)
print('True negative rate at k =4: {:.2f}'.format(TNR))

#Precision (Class 1)
precision = precision_score(y_test, ypred_test_optimal)
print ('Precision: {:.2f}'.format(precision))

#F1 Score 
f1 = f1_score(y_test, ypred_test_optimal)
print('F1-Score: {:.2f}'.format(f1))
```

2)B) Learning Curve
-

```python
N_subset = list(range(10,211,10)) #create training set sizes, N: 10,20, 30...210
test_errors_optimal =[] #list that will store best test error at each N


for N in N_subset: 

    class0_training_subset = class0.iloc[: int(N/3)]
    class1_training_subset = class1.iloc[: int(N - N/3)]
    training_subset = pd.concat([class0_training_subset, class1_training_subset]) 

    #Now for each value of N, we need to create a set of k: 
    #For example, if N = 200, the optimal k is selected from  {1,6, 11,..196}, thus k < N

    k_subset = list(range(1,N,5))

    x_training_subset = training_subset[['pelvic_incidence', 'pelvic_tilt', 'lumbar_lordosis_angle','sacral_slope', 'pelvic_radius', 'degree_spondylolisthesis']]
    y_training_subset = training_subset['class']

    test_errors_subset = []
    
    for k in k_subset: 
        if k < len(training_subset): #Ensure k is valid relative to training set size
            knn = KNeighborsClassifier(n_neighbors = k, metric = 'euclidean')
            knn.fit(x_training_subset, y_training_subset)
            #Calculate errors on test data. We use accuracy_score and not MSE since this is a classification problem. 
            ypred_test_subset = knn.predict(x_test)
            test_errors_subset.append(1-accuracy_score(y_test, ypred_test_subset))
            
            
    #Now for each N, we need to find the lowest k and append to master list
    test_errors_optimal.append(min(test_errors_subset))


plt.figure(figsize=(15,8))
plt.plot(N_subset, test_errors_optimal, 'o', linestyle = '-',color = 'b')
plt.xlabel('Training set size (N)', fontsize = 14)
plt.ylabel('Minimum test error for each N', fontsize = 14)
plt.title('Learning Curve')
plt.grid(True)
plt.show()
```

Manhattan Distance
-

```python
manhattan_errors = []
manhattan_error_optimal = 1
manhattan_k_optimal = 1

k_set = list(range(1,197,5))
for k in k_set:
    knn = KNeighborsClassifier(n_neighbors = k, metric = 'manhattan')
    knn.fit(x_training, y_training)
    ypred_test = knn.predict(x_test)
    test_error = 1-accuracy_score(y_test, ypred_test)
    manhattan_errors.append(test_error)
    if test_error < manhattan_error_optimal: 
        manhattan_error_optimal = test_error
        manhattan_k_optimal = k

#Plot Manhattan test errors
plt.figure(figsize=(15,8))
plt.plot(k_set, manhattan_errors,'b',label = 'Manhattan Test Error')
plt.xlabel('k value')
plt.title('Plotting errors vs. k values')
plt.legend(loc = 'upper left')
plt.grid(True)
plt.show()
        
```

```python
# Summarize Manhattan optimal k
print ("Lowest Manhattan error: ", round(manhattan_error_optimal,2))
print ("Corresponding to k = ", manhattan_k_optimal)
```

Minkowski Distance
-
 Next, we will calculate Minkowski Distance with log(p) in {0.1. 0.2,..1.0}
We will use k = 6 as we got from Manhattan Distance 


```python
logp_set = np.arange(0.1,1.1,0.1)
minkowski_errors = {}
minkowski_error_optimal = 1
minkowski_logp_optimal = 0.1

for logp in logp_set:
    pval = 10**logp
    knn = KNeighborsClassifier(n_neighbors = manhattan_k_optimal, metric = 'minkowski', p = pval)
    knn.fit(x_training, y_training)
    ypred_test = knn.predict(x_test)
    test_error = 1 - accuracy_score(y_test, ypred_test)
    minkowski_errors[float(round(logp,1))] = round(test_error,2)
    if test_error < minkowski_error_optimal:
        minkowski_error_optimal = round(test_error,2)
        minkowski_logp_optimal = float(round(logp,1))

print("log10(p):  ", "errors")
for logp, error in minkowski_errors.items():
    print(logp, error, sep = '          ')
```

best log(p) is 0.6 with error rate of 0.06

Mahalanobis Distance
-

```python
#Computing Covariance Matrix:

cov_matrix = np.cov(x_training, rowvar = False)

# Check conditions: determinant, rank, condition number, and eigenvalues
determinant = np.linalg.det(cov_matrix)
rank = np.linalg.matrix_rank(cov_matrix)
condition_number = np.linalg.cond(cov_matrix)
eigenvalues = np.linalg.eigvals(cov_matrix)


# Use pseudoinverse if matrix is singular or ill-conditioned
if np.linalg.cond(cov_matrix) > 1e10 or np.isclose(np.linalg.det(cov_matrix), 0):
    inv_cov_matrix = np.linalg.pinv(cov_matrix)
else:
    inv_cov_matrix = np.linalg.inv(cov_matrix)

mahalanobis_errors = []
mahalanobis_error_optimal = 1
mahalanobis_k_optimal = 1

for k in k_set:
    knn = KNeighborsClassifier(n_neighbors = k, metric='mahalanobis', metric_params={'VI': inv_cov_matrix})
    knn.fit(x_training, y_training)
    ypred_test = knn.predict(x_test)
    test_error = 1-accuracy_score(y_test, ypred_test)
    mahalanobis_errors.append(test_error)
    if test_error < mahalanobis_error_optimal: 
        mahalanobis_error_optimal = test_error
        mahalanobis_k_optimal = k

#Plot Chebyshev test errors
plt.figure(figsize=(15,8))
plt.plot(k_set, mahalanobis_errors,'b',label = 'Mahalanobis Test Error')
plt.xlabel('k value')
plt.title('Plotting Mahalanobis errors vs. k values')
plt.legend(loc = 'upper left')
plt.grid(True)
plt.show()

    
```

```python

#First Calculate Euclidean weighted optimal K and error rate 
weight_euclidean_k_optimal = 1
weight_euclidean_error_optimal = 1

#Iterate through k values
for k in k_set:
    knn = KNeighborsClassifier(n_neighbors=k, metric = 'euclidean', weights = 'distance')
    knn.fit(x_training, y_training)

    ypred_test = knn.predict(x_test)
    test_error = 1-accuracy_score(y_test, ypred_test)
    if test_error < weight_euclidean_error_optimal:
        weight_euclidean_error_optimal = test_error
        weight_euclidean_k_optimal = k

#Calculate Manhattan optimal values using weighted distances
weight_manhattan_error_optimal = 1
weight_manhattan_k_optimal = 1

for k in k_set:
    knn = KNeighborsClassifier(n_neighbors = k, metric = 'manhattan', weights = 'distance')
    knn.fit(x_training, y_training)
    ypred_test = knn.predict(x_test)
    test_error = 1-accuracy_score(y_test, ypred_test)
    if test_error < weight_manhattan_error_optimal: 
        weight_manhattan_error_optimal = test_error
        weight_manhattan_k_optimal = k

#Calculate Chebyshev error and optimal k using weighted distances
weight_chebyshev_error_optimal = 1
weight_chebyshev_k_optimal = 1
for k in k_set:
    knn = KNeighborsClassifier(n_neighbors = k, metric = 'chebyshev',weights = 'distance')
    knn.fit(x_training, y_training)
    ypred_test = knn.predict(x_test)
    test_error = 1-accuracy_score(y_test, ypred_test)
    if test_error < weight_chebyshev_error_optimal: 
        weight_chebyshev_error_optimal = test_error
        weight_chebyshev_k_optimal = k

weighted_results_summary = {
     'Distance Metric' : ['Euclidean','Manhattan', 'Chebyshev'], 
     'Optimal k: ': [weight_euclidean_k_optimal, weight_manhattan_k_optimal, weight_chebyshev_k_optimal],
    'Error rate: ': [round(weight_euclidean_error_optimal, 2), round(weight_manhattan_error_optimal, 2), round(weight_chebyshev_error_optimal, 2)]
}
weighted_summary = pd.DataFrame(weighted_results_summary)
print(weighted_summary)
```
